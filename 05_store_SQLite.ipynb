```python
import pandas as pd 
final_df = pd.read_csv("data/WHO_WorldBank_merged.csv")
```

```python
import sqlite3
import pandas as pd

# Connect (creates the file if it doesn't exist)
conn = sqlite3.connect("data/db/mental_health.db")

# Write the DataFrame to SQLite
final_df.to_sql(
    "mental_health_data",  # table name
    conn,
    if_exists="replace",   # replace if already exists
    index=False
)

print("Data saved to SQLite successfully!")
conn.close()
```

```python
#Countries with high mental health values in 2010.

conn = sqlite3.connect("data/db/mental_health.db")
query = """
SELECT country_name_y AS country,
       year,
       mental_health_value,
       GDP_per_capita,
       Unemployment
FROM mental_health_data
WHERE year = 2010 AND mental_health_value > 20
ORDER BY mental_health_value DESC;
"""

df = pd.read_sql(query, conn)
conn.close()

print(df.head())
```

```python
# Countries with the highest GDP per capita in 2020.

conn = sqlite3.connect("data/db/mental_health.db")

query2 = """
SELECT country_name_y AS country,
       year,
       GDP_per_capita,
       mental_health_value
FROM mental_health_data
WHERE year = 2020
ORDER BY GDP_per_capita DESC
LIMIT 5;
"""

top_gdp_df = pd.read_sql(query2, conn)
conn.close()

print(top_gdp_df)
```
