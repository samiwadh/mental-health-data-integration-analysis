```python
# Load raw data another dataaset 
wb_df = pd.read_csv("data//raw//world_bank_data_raw.csv")
wb_df
```

```python

# WB raw data
print("WB missing values", wb_df.isna().sum())
```

```python
# Fill missing values for each column
for col in ["GDP_per_capita", "School_enrollment", "Unemployment"]:
    wb_df[col] = wb_df[col].fillna(wb_df[col].mean())
```

```python
wb_df.isna().sum()
```

```python
# Save cleaned dataset
wb_df.to_csv("data//processed//world_bank_data_clean.csv", index=False)
```
